```python
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents=["Gangs of Wasseypur is a great movie,", "The success of a movie depends upon the performance of the actors.","There are no new movies releasing this week."]
print(documents)
```

```python
def preprocess(document):
    document=document.lower()
    words=word_tokenize(document)
    words=[word for word in words if word not in stopwords.words("english")]
    document=" ".join(words)
    return document


documents = [preprocess(document) for document in documents]
print(documents)
```

# Create bag of words model using count vectorizer function

```python
vectorizer=CountVectorizer()
bow_model=vectorizer.fit_transform(documents)
print(bow_model)
```

```python
print(bow_model.toarray())
```

```python
print(bow_model.shape)
print(vectorizer.get_feature_names())
```

# create a bag of words model on spam dataset

```python
spam=pd.read_csv("SMSSpamCollection.txt",sep="\t" ,names=["label","message"])
spam.head()
```

```python
spam=spam.iloc[0:50,:]
print(spam)
```

```python
messages=spam.message
print(messages)
```

```python
messages=[message for message in messages]
print(messages)
```

```python
messages=[preprocess(message) for message in messages]
print(messages)
```

```python
vectorizer=CountVectorizer()
bow_model=vectorizer.fit_transform(messages)
```

```python
pd.DataFrame(bow_model.toarray(),columns=vectorizer.get_feature_names())
```
